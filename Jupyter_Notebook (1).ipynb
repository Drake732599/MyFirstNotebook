### This code, when executed, plots a sine wave with an amplitude of 1.

```python
import numpy as np
import matplotlib.pyplot as plot
value = np.arange(0, 10, 0.01);
amplitude = np.sin(value)
plot.plot(value, amplitude)
plot.title('Sine')
plot.xlabel('Value')
plot.ylabel('Amplitude = sin(value)')
plot.axhline(y=0, color='r')
plot.show()
plot.show()
```
